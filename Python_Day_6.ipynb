## Tuple in Python

A tuple is an ordered, immutable collection in Python. It can store a sequence of heterogeneous (different) data types. Once a tuple is created, its contents cannot be changed (unlike lists).

```python
# tuple is immutable / unchangeable vs list is mutable changeable L = []
```

```python
#### Tuple Creation:

t1 = (1, 2, 3)
t2 = ()               # empty tuple
t3 = (5)             # single-element tuple (note the comma!)
t4 = tuple("abc")     # from iterable: ('a', 'b', 'c')
```

```python
print(t1,type(t1))
```

```python
t2
```

```python
t3 = (5,)             # single-element tuple (note the comma!)
print(t3,type(t3))
```

```python
t4
```

```python
# Homogenous Data Types:

tup2 = (10,30,60) # tuple of integers numbers
tup3 = (10.77,30.66,60.89) # tuple of float numbers
tup4 = ('one','two' , "three") # tuple of strings
```

## Set in Python
In Python, a set is an unordered, mutable, and unindexed collection of unique elements. It's useful when you want to store items without duplicates.



```python
#### creating a Set:

# Using curly braces
my_set = {1, 2, 3, 4}
```

```python
# Using the set() constructor
another_set = set([4, 5, 6])
```

```python
my_set
```

```python
another_set
```

```python
len(my_set) # length of the set
```

```python
my_set = {1,1,2,2,3,4,5,5}
my_set # Duplicate elements are not allowed
```

```python
myset1 = {1.79,2.08,3.99,4.56,5.45} # Set of float numbers
myset1
```

```python
myset2 = {'Asif' , 'John' , 'Tyrion'} # Set of Strings
myset2
```

```python
myset3 = {10,20, "Hola", (11, 22, 32)} # Mixed datatypes
myset3
```

```python
myset3 = {10,20, "Hola", [11, 22, 32]} # set doesn't allow mutable items like list
myset3
```

```python
my_set1 = set(('one' , 'two' , 'three' , 'four'))
my_set1
```

#### Add & Remove Items in set

```python
myset = {'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight'}
```

```python
myset.add('NINE') # Add item to a set using add() method
myset
```

```python
myset.update(['TEN' , 'ELEVEN' , 'TWELVE']) # Add multiple item to a set using update
myset
```

```python
myset.remove('NINE') # remove item in a set using remove() method
myset
```

```python
myset.discard('TEN') # remove item from a set using discard() method
myset
```

```python
# Difference between remove () and Discard()

my_set = {1, 2, 3}

my_set.add(4)             # Adds an element
my_set.update([5, 6])     # Adds multiple elements
my_set.remove(2)          # Removes element; error if not found
my_set.discard(10)        # Removes element; no error if not found
```

```python
my_set
```

##### Clear()

```python
my_set.clear()            # Removes all elements
```

```python
# Deleting a set:

del my_set
```

```python
my_set
```

#### Set Operation


```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

#### Union():

```python
# Union:

A | B # Union of A and B (All elements from both sets. NO DUPLICATES)
```

```python
A.union(B) # Union of A and B
```

```python
A.union(B) # Union of A and B
```

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}

# update()  # perform union operation and the result should be updated in 'A' Set itself

A.update(B,C)
A
```

#### Intersection()


```python
# Intersection()

A= {1,2,3,4,5}
B = {4,5,6,7,8}

A & B # Intersection of A and B (Common items in both sets)
```

```python
A.intersection(B) # Intersection of A and B
```

```python
A
```

```python
A.intersection_update(B)
A
```

#### Difference:


```python
# Difference:

A= {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A - B # set of elements that are only in A but not in B
```

```python
A.difference(B) # Difference of sets
```

```python
B- A # set of elements that are only in B but not in A
```

```python
B.difference(A)
```

####  Symmetric Difference


```python
A= {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A ^ B # Symmetric difference (Set of elements in A and B but not in both EXCLUDE COMMON element)
```

```python
A.symmetric_difference(B) # Symmetric difference of sets
```

```python
A
```

```python
A.symmetric_difference_update(B)
A
```

## Dictionary in Python
- Dictionary is a mutable data type in Python.
- A python dictionary is a collection of key and value pairs separated by a colon (:) & enclosed
  in curly braces {}.
- Keys must be unique in a dictionary, duplicate values are allowed.

#### Create Dictionary:

```python
mydict = dict() # empty dictionary
mydict
```

```python
mydict = {} # empty dictionary
mydict
```

```python
mydict = {1:'one' , 2:'two' , 3:'three'} # dictionary with integer keys
mydict
```

```python
mydict = dict({1:'one' , 2:'two' , 3:'three'}) # Create dictionary using dict()
mydict
```

```python
mydict = {'A':'one' , 'B':'two' , 'C':'three'} # dictionary with character keys
mydict
```

```python
mydict.keys() # Return Dictionary Keys using keys() method
```

```python
mydict.values() # Return Dictionary Values using values() method
```

```python
mydict.items() # Access each key-value pair within a dictionary
```

```python
mydict = {1:'one' , 2:'two' , 'A':['asif' , 'john' , 'Maria']} # dictionary using list as values
mydict
```

```python
mydict = {1:'one' , 2:'two' , 'A':['asif' , 'john' , 'Maria'], 'B':('Bat' , 'cat','rat')}
mydict
```
